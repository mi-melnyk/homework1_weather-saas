```python
import requests
import json
```

```python
RSA_KEY = ""
GEMINI_KEY = ""
```

```python
payload = {
    "token": "***",
    "requester_name": "Milana Melnyk",
    "location": "Kyiv, Ukraine",
    "date": "2026-02-01"
}
```

```python
city = payload["location"]
date = payload["date"]

url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{city}/{date}?unitGroup=metric&key={RSA_KEY}&contentType=json"

response = requests.get(url)
```

```python
if response.status_code == requests.codes.ok:
    print(response.text)
else:
    print("Error:", response.status_code, response.text)
```

```python
response_json = json.loads(response.text)
response_json
```

```python
print(json.dumps(response_json, indent=4))
```

```python
temp = response_json['days'][0]['temp']
wind = response_json['days'][0]['windspeed']
print(f"Temperature: {temp}°C, Wind: {wind} kph")
```

```python
final_response = {
    "requester_name": "Milana Melnyk",
    "timestamp": "2026-02-21T19:00:00Z",
    "location": city,
    "date": date,
    "weather": {
        "temp_c": response_json['days'][0]['temp'],
        "wind_kph": response_json['days'][0]['windspeed'],
        "pressure_mb": response_json['days'][0]['pressure'],
        "humidity": response_json['days'][0]['humidity']
    }
}
print(json.dumps(final_response, indent=4))
```

```python
precip = response_json['days'][0].get('precip', 0)
precip_prob = response_json['days'][0].get('precipprob', 0)

precip_desc = f"precip: {precip} mm, probability: {precip_prob}%"
url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash:generateContent?key={GEMINI_KEY}"
headers = {"Content-Type": "application/json"}

prompt_text = (
    f"Current weather in {city} on {date}: {temp}°C, wind {wind} kph, {precip_desc}. "
    f"Suggest one creative evening activity (max 30 words). "
    f"Don't limit yourself to specific options, suggest anything relevant to the weather."
)

payload = {
    "contents": [{
        "parts": [{"text": prompt_text}]
    }]
}

response_ai = requests.post(url, headers=headers, json=payload)
```

```python
def get_recommendation(res):
    if res.status_code == 200:
        data = res.json()
        try:
            return data["candidates"][0]["content"]["parts"][0]["text"].strip()
        except (KeyError, IndexError):
            return "Enjoy your evening regardless of the weather!"
    elif res.status_code == 429:
        return "AI is busy (API limit reached). Please wait a minute."
    else:
        return f"AI Error {res.status_code}: {res.text}"
```

```python
recommendation = get_recommendation(response_ai)
print(f"Location: {city}")
print(f"Weather: {temp}°C, {response_json['days'][0]['conditions']}")
print(f"AI Recommendation: {recommendation}")
```
